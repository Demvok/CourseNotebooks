```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
%matplotlib inline
root = pd.read_csv('iris.csv').iloc[:, 1:]
root
```

```python
iris_type = tree.DecisionTreeClassifier(criterion='entropy')
x_axis = root.iloc[:, :4]
y_axis = root.iloc[:, 4]

iris_type.fit(X=x_axis, y=y_axis)
```

```python
tree.plot_tree(iris_type, feature_names=root.columns)
```

```python
sns.pairplot(root,hue='species')
```

```python
sns.violinplot(root['petal length'])
```

```python
sns.distplot(root['sepal length'], label='Sepal length')
sns.distplot(root['sepal width'], label='Sepal width')
sns.distplot(root['petal length'], label='Petal length')
sns.distplot(root['petal width'], label='Petal width')
plt.legend()
```
